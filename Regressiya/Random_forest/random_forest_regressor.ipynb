```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
np.random.seed(1)
X = np.sort(np.random.randint(30, 150, 50)).reshape(-1, 1)
y = (X.ravel() * 1.2 + np.random.normal(0, 10, X.shape[0]))
```

```python
rf = RandomForestRegressor(n_estimators=100, random_state=1)
```

```python
rf.fit(X, y)
```

```python
y_pred = rf.predict(X)
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(X, y, color='gray', label='uy narxlari')
plt.plot(X, y_pred, color='green', label='Random Forest Regression')
plt.xlabel("Uy oâ€˜lchami (kv.m)")
plt.ylabel("Uy narxi (ming $)")
plt.title("Random Forest Regression: Uy narxini bashorat qilish")
plt.legend()
plt.grid(True)
plt.show()
```
