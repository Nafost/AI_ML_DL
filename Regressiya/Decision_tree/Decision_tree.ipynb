```python
# import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# import seaborn as sns
from sklearn.model_selection import train_test_split
# from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
data = {
    'yoshi': [22, 25, 19, 35, 40, 50, 28],
    'oylik_daromadi_$': [600, 800, 500, 2500, 3200,3800,1200]
}

df = pd.DataFrame(data)
```

```python
df
```

```python
X = df[['yoshi']]  
y = df['oylik_daromadi_$']
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import plot_tree
```

```python
tree = DecisionTreeRegressor()  
tree.fit(X, y)
```

```python
yangi_yosh = pd.DataFrame({'yoshi': [32]})
```

```python
predicted_sales = tree.predict(yangi_yosh)
```

```python
print("32 yosh uchun oylik daromad:", predicted_sales[0])
```

```python
df
```

```python
plt.figure(figsize=(8, 6))  
plot_tree(tree, feature_names=X.columns, filled=True, rounded=True)
plt.title("Turli yoshlar uchun oylik daromad")
plt.show()
```

```python
data = {
    'ish_soati': [12, 8, 10, 6, 16],
    'ish_tajribasi': [3, 2, 3, 2, 4],
    'oylik_maoshi_$': [3500, 2800, 3000, 2500, 4000]
}

df = pd.DataFrame(data)
```

```python
df
```

```python
X = df[['ish_soati', 'ish_tajribasi']]  
y = df['oylik_maoshi_$'] 
```

```python
tree = DecisionTreeRegressor(max_depth=10)  
tree.fit(X, y)
```

```python
yangi_ishchi = pd.DataFrame({
    'ish_soati': [14],
    'ish_tajribasi': [3]
})
```

```python
predicted_price = tree.predict(yangi_ishchi)
```

```python
print("yangi ishchi uchun oylik maosh:", predicted_price[0])
```

```python
plt.figure(figsize=(12, 8)) 
plot_tree(tree, feature_names=X.columns, filled=True, rounded=True)
plt.title("ishchilar oyligi uchun qarorlar daraxti")
plt.show()
```
