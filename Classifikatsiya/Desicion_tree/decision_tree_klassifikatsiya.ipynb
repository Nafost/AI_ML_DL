```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.metrics import accuracy_score
```

```python
np.random.seed(1)
exam1 = np.random.randint(30, 100, 100)
exam2 = np.random.randint(30, 100, 100)
passed = ((exam1 + exam2)/2 > 60).astype(int)
```

```python
X = np.column_stack((exam1, exam2))
y = passed
```

```python
model = DecisionTreeClassifier()
model.fit(X, y)
```

```python
plt.figure(figsize=(6,6))
plt.scatter(exam1, exam2, c=y, cmap='coolwarm', edgecolors='k')
plt.xlabel("Imtihon 1")
plt.ylabel("Imtihon 2")
plt.title("Decision Tree: Oâ€˜tgan yoki yoâ€˜qligini klassifikatsiya qilish")
plt.grid(True)
plt.show()
```
